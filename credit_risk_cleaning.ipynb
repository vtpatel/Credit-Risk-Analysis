# CREDIT RISK ANALYSIS - LENDING CLUB LOAN DATA
# DATA CLEANING


# Load Dataset


```python
# import libraries

import pandas as pd
import numpy as np

df=pd.read_csv('/kaggle/input/datasets/wordsforthewise/lending-club/accepted_2007_to_2018Q4.csv.gz', low_memory=False)

# Print first 5 rows

df.head()
```

# Explore Dataset

```python
df.info()

# check shape of the dataset

print("Rows  :", df.shape[0])
print("Columns  :", df.shape[1])
```

# Removes duplicates


```python
df.duplicated().sum()
```

# Handling Missing Values


```python
df.isnull().sum()
```

```python
# filling missing numerical values

df.fillna(df.median(numeric_only=True), inplace=True) 
```

```python
# filling missing categorial values

df.fillna(df.mode().iloc[0],inplace=True)

df.isnull().sum() 

# drop member_id column as most of the values are NaN

df=df.drop('member_id', axis=1)
```

```python
# Print all column names one by one

for col in df.columns:
    print(col)
```

# Clean Loan Status Column


```python
# check values in loan status column

df['loan_status'].value_counts()
```

```python
#keeping only imp loan status value

df = df[df['loan_status'].isin(['Fully Paid', 'Charged Off'])]
#convert loan status into numbers

df['loan_status'] = df['loan_status'].map({'Fully Paid': 0, 'Charged Off': 1})
```

```python
df['loan_status'].unique()
```

# Created a New Feature


```python
# Create a loan to income ratio column
df['loan_to_income']=df['loan_amnt']/df['annual_inc']
```

```python
#check new column
df[['loan_amnt','annual_inc','loan_to_income']].head()

#check remaining missing values

df.isnull().sum()

```

```python
# Keep only required columns

df = df[[
    'id',
    'loan_amnt',
    'funded_amnt',
    'term',
    'int_rate',
    'installment',
    'grade',
    'sub_grade',
    'emp_length',
    'home_ownership',
    'annual_inc',
    'verification_status',
    'loan_status',
    'purpose',
    'dti',
    'fico_range_low',
    'fico_range_high',
    'open_acc',
    'pub_rec',
    'revol_bal',
    'revol_util',
    'total_acc',
    'application_type',
    'loan_to_income'
]]
```

```python
df.tail()

# removing last two rows

df=df[:-2]

# there are rows in id columns which contains text, so we are only keeping numeric ids to avoid error

# Keep only numeric id values

df = df[df['id'].astype(str).str.isnumeric()]
```

```python
# Saved cleaned credit risk dataset

df.to_csv('cleaned_credit_risk_dataset.csv', index=False)
```
